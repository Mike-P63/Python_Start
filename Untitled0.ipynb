Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)


```python
penguins.loc[penguins["bill_length_mm"]>=42,"i_bill"] = "High"
penguins.loc[(penguins["bill_length_mm"]>=35)&(penguins["bill_length_mm"]<=42),"i_bill"] = "Middle"
penguins.loc[penguins["bill_length_mm"]<=35,"i_bill"] = "Low"
penguins.head(10)
```

Изобразить гистограмму по flipper_length_mm с оттенком height_group. Сделать анализ

```python
sns.histplot(data=penguins,x="flipper_length_mm", hue="i_bill",multiple='dodge')
```
